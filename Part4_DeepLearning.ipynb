```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv")
```

```python
df["Age"]=df["Age"].astype("float")
```

```python
x1=df[['Cement', 'Blast Furnace Slag', 'Fly Ash', 'Water', 'Superplasticizer',
       'Coarse Aggregate', 'Fine Aggregate', 'Age']]
y=df["Strength"]
```

```python
x1=(x1-x1.mean()/x1.std())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(x1, y, test_size=0.3, random_state=4)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import mean_squared_error
```

```python
def regression_mod():
    model=Sequential()
    number_input=x1.shape[1]

    model.add(Dense(10, activation="relu", input_shape=((number_input,))))
    model.add(Dense(10, activation="relu"))
    model.add(Dense(10, activation="relu"))
    model.add(Dense(1, activation="relu"))

    model.compile(optimizer="adam", loss="mean_squared_error")
    model.fit(X_train, y_train, epochs=100)
    modelhat=model.predict(X_test)
    
    return modelhat
```

```python
mse=[]
```

```python
for i in range(0,50):
    mse.append(mean_squared_error(y_test,regression_mod() ))
```

```python
len(mse)
```

```python
mse
```

```python
Mean_Part4=sum(mse)/len(mse)
variance = sum([((i - Mean_Part4) ** 2) for i in mse]) / len(mse)
Std_Part4= variance ** 0.5
```

```python
print("Mean for the Part4 ", Mean_Part4)
print("STD for the Part 4 ", Std_Part4) 
```

We can see that the Mean of the mean squared error has been reduced when compared to the part B and C
